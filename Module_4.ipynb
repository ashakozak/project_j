```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# File to Load
international_port = Path("Resources/internationalport_country.csv")
local_city = Path("Resources/localcity_australia.csv")

# Read airline Data File and store into Pandas DataFrames
international_port_df = pd.read_csv(international_port)
local_city_df = pd.read_csv(local_city)
```

```python
# Display the data table for international_port
international_port_df
```

```python
# Get data after 2000 from international_port_df
international_year = international_port_df.loc[international_port_df["Year"]>=2000]
international_year
```

```python
# Drop rows with zero passengers
international_year_dropped = international_year.drop(international_year[international_year["Passengers_In"] ==0].index)
international_year_dropped
```

```python
# Replace month_num with months
months_dict = {1: "January", 2: "February", 3: "March", 4: "April", 5: "May", 6: "June", 7: "July", 8: "August",
               9: "September", 10: "October", 11: "November", 12: "December"}
international_year_dropped["Month_num"] = international_year_dropped["Month_num"].replace(months_dict)
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September",
         "October", "November", "December"]
international_year_dropped["Month_num"] = pd.Categorical(international_year_dropped["Month_num"], categories=months, ordered=True)
```

```python
# Identify the travel months
months_travel = international_year_dropped["Month_num"].unique().tolist()
months_travel
```

```python
# Identify busiest season by determining total number of passengers from international_year_dropped - groupby month
international_monthly_passenger = international_year_dropped.groupby("Month_num")[["Passengers_In"]].mean()
international_monthly_passenger
```

```python
# Pyplot busiest season from 2000 - 2022

plt.figure(figsize=(9,7))
plt.bar(international_monthly_passenger.index,international_monthly_passenger["Passengers_In"])
plt.xlabel("Months")
plt.xticks(rotation=45)
plt.ylabel("Average Number of Travellers")
plt.title("Busiest Season of International Travellers to Australia 2000 - 2022")
plt.grid()
plt.savefig('busiest_season.png')
plt.tight_layout

plt.show()
```

```python
# Identify travel trend of busiest season from 2000 - 2022
international_monthly_pax_yr = international_year_dropped.groupby(["Year", "Month_num"])[["Passengers_In"]].mean()
international_monthly_pax_yr

# Reset index
international_monthly_pax_yr = international_monthly_pax_yr.reset_index()
international_monthly_pax_yr
```

```python
# Create dataframe for passengers per month from 2000 - 2022
january_pax = international_monthly_pax_yr.loc[international_monthly_pax_yr["Month_num"] == "January",
                                              ["Year", "Month_num", "Passengers_In"]]
february_pax = international_monthly_pax_yr.loc[international_monthly_pax_yr["Month_num"] == "February",
                                              ["Year", "Month_num", "Passengers_In"]]
march_pax = international_monthly_pax_yr.loc[international_monthly_pax_yr["Month_num"] == "March",
                                              ["Year", "Month_num", "Passengers_In"]]
april_pax = international_monthly_pax_yr.loc[international_monthly_pax_yr["Month_num"] == "April",
                                              ["Year", "Month_num", "Passengers_In"]]
may_pax = international_monthly_pax_yr.loc[international_monthly_pax_yr["Month_num"] == "May",
                                              ["Year", "Month_num", "Passengers_In"]]
june_pax = international_monthly_pax_yr.loc[international_monthly_pax_yr["Month_num"] == "June",
                                              ["Year", "Month_num", "Passengers_In"]]
july_pax = international_monthly_pax_yr.loc[international_monthly_pax_yr["Month_num"] == "July",
                                              ["Year", "Month_num", "Passengers_In"]]
august_pax = international_monthly_pax_yr.loc[international_monthly_pax_yr["Month_num"] == "August",
                                              ["Year", "Month_num", "Passengers_In"]]
september_pax = international_monthly_pax_yr.loc[international_monthly_pax_yr["Month_num"] == "September",
                                              ["Year", "Month_num", "Passengers_In"]]
october_pax = international_monthly_pax_yr.loc[international_monthly_pax_yr["Month_num"] == "October",
                                              ["Year", "Month_num", "Passengers_In"]]
november_pax = international_monthly_pax_yr.loc[international_monthly_pax_yr["Month_num"] == "November",
                                              ["Year", "Month_num", "Passengers_In"]]
december_pax = international_monthly_pax_yr.loc[international_monthly_pax_yr["Month_num"] == "December",
                                              ["Year", "Month_num", "Passengers_In"]]
```

```python
# Pyplot travel trend of busiest season from 2000 - 2022

plt.figure(figsize=(10,7))
plt.plot(january_pax["Year"], january_pax["Passengers_In"], label = "January")
plt.plot(february_pax["Year"], february_pax["Passengers_In"], label = "February")
plt.plot(march_pax["Year"], march_pax["Passengers_In"], label = "March")
plt.plot(april_pax["Year"], april_pax["Passengers_In"], label = "April")
plt.plot(may_pax["Year"], may_pax["Passengers_In"], label = "May")
plt.plot(june_pax["Year"], june_pax["Passengers_In"], label = "June")
plt.plot(july_pax["Year"], july_pax["Passengers_In"], label = "July")
plt.plot(august_pax["Year"], august_pax["Passengers_In"], label = "August")
plt.plot(september_pax["Year"], september_pax["Passengers_In"], label = "September")
plt.plot(october_pax["Year"], october_pax["Passengers_In"], label = "October")
plt.plot(november_pax["Year"], november_pax["Passengers_In"], label = "November")
plt.plot(december_pax["Year"], december_pax["Passengers_In"], label = "December")
plt.title("Busiest Season Throughout the Years 2000 - 2022")
plt.xlabel("Years")
plt.ylabel("Average Numbers of Travellers")
plt.grid()
plt.legend(loc="best")
plt.savefig('season_years.png')
plt.tight_layout

plt.show()
```

ANALYSIS:

The busiest season of international travellers to Australia are during the months of January, July, October and December.
January - with the Australian Open, and the start of the new school year.
July - winter, or the good weather to visit the beaches of Sydney and Brisbane.
October - spring, and Melbourne Cup.
December - holiday and with overseas returning to celebrate with family.
These months have consistently been the busiest season throughout the years.
The most visited Australian Cities are Sydney, Melbourne, Perth and Gold Coast.
The busiest season of these Australian cities are the months of June, July, October and December.
The decline of international travellers from 2020 can be significantly attributed to Covid 19, with the closure of international border.
